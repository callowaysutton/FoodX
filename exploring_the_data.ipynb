# Visualizing the Data
Hmm, those are definitely some statistcs... but it's really not that useful, especially since a lot of these fields are strings (so sad). But now I can try to visualizing it to see if there are any potential trends within the data

```python
import matplotlib.pyplot as plt

# Group data by university and count orders
university_orders = df['University'].value_counts()

# Plot the distribution of orders by university
plt.figure(figsize=(12, 6))
plt.bar(university_orders.index, university_orders.values)
plt.xticks(rotation=90)
plt.xlabel('University')
plt.ylabel('Number of Orders')
plt.title('Distribution of Orders by University')
plt.show()
```

```python
# Group data by food item and count orders
food_orders = df['Order'].value_counts().head(10)

# Plot the most popular food items
plt.figure(figsize=(10, 6))
food_orders.plot(kind='bar')
plt.xlabel('Food Item')
plt.ylabel('Number of Orders')
plt.title('Top 10 Most Popular Food Items')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('FoodX.csv')

# Group data by student year and count orders
year_orders = df['Year'].value_counts().sort_index()

# Plot the orders by student year
plt.figure(figsize=(8, 5))
year_orders.plot(kind='bar', color='skyblue')
plt.xlabel('Student Year')
plt.ylabel('Number of Orders')
plt.title('Orders by Student Year')
plt.xticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('FoodX.csv')

# Create a histogram for the 'Time' column
plt.figure(figsize=(10, 6))
sns.histplot(data['Time'], bins=24, kde=True)
plt.title('Distribution of Order Times')
plt.xlabel('Time (Hour of Day)')
plt.ylabel('Frequency')
plt.xticks(range(24))  # Set x-axis ticks to show hours
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add grid lines for better readability
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('FoodX.csv')

# Group data by major and count orders
major_orders = df['Major'].value_counts()

# Sort the major_orders Series by the count of orders in descending order
major_orders = major_orders.sort_values(ascending=False)

# Create a bar chart to visualize orders by major
plt.figure(figsize=(12, 6))
major_orders.plot(kind='bar')
plt.xlabel('Major')
plt.ylabel('Number of Orders')
plt.title('Distribution of Orders by Student Major')
plt.xticks(rotation=90)
plt.show()
```
